```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Cleaned_Dataframe.xlsx')
df.set_index('Sample',inplace=True)
```

```python
df_control = df.loc[df['Status'] == 'Control']
df1 = df_control[(df_control['Age at sample'] > 20) & 
                   (df_control['Age at sample'] < 60)]
```

```python
df1
```

```python
x = df1['Age at sample']
y = df1['GP1']
plt.plot(x, y, 'o')
m1, b = np.polyfit(x, y, 1)
plt.plot(x, m1*x + b)
plt.xlabel('Age at sample')
plt.ylabel('GP1')
print(m1)

matrix1 = np.corrcoef(x, y)
r1 = matrix1[0,1]**2
print(r1)
```
